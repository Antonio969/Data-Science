# Visualizacion de datos con matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# ejemplo de una linea simple 
x = [0,1,2,3,4,5]
y = [0,1,2,3,4,5]

plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Mi primera grafica')


```

### Mas graficos

```python
# Grafico de barras
categorias = ['A','B','C','D']
valores = [10,20,30,40]
plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Grafica de barras')
```

```python
plt.style.use('ggplot')
plt.style.available
```

```python
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]

plt.plot(x,y, color = 'red', linestyle = '--', marker='o')
plt.xlabel('Tiempo')
plt.ylabel('Valor medido')
```
