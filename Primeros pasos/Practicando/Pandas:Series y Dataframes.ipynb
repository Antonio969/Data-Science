### Agrupacion de datos

```python
data = {
    'Nombre': ['Juan','Ana','Luis','Laura','Pedro','Carla'],
    'Ciudad': ['Madrid','Barcelona','Madrid','Valencia','Barcelona','Madrid'],
    'Edad':[25,33,30,28,45,38],
    'Puntuacion':[80,90,85,88,75,91]
}
df = pd.DataFrame(data)
df
```

```python
# Agrupar los datos por ciudad
grouped = df.groupby('Ciudad')
print(grouped.groups)
```

La aregacion de datos es el proceso de combinar multiples filas de datos en una sola fila, aplicando una funcion de agregacion como suma, promedio, minimo max, etc.

```python
# Calcular la suma de las edades y puntuacion por ciudad
aggragated_data = grouped.agg(
    {
        'Edad': 'mean',
        'Puntuacion': 'sum'
    }
)
aggragated_data
```

```python
# Definir una funcion de agregacion personalizada
def rango(serie):
    return serie.max()-serie.min()

aggragated_data_custom = grouped.agg(
    {
        'Edad': rango,
        'Puntuacion': rango
    }
)
aggragated_data_custom
```

```python
data['Categoria'] = ['A','B','A','B','A','B']
df = pd.DataFrame(data)
df
```

```python
# Agrupar datos por ciudad y categoria
grouped_multi = df.groupby(['Ciudad','Categoria'])
grouped_multi.groups
```

```python
# Calcular la suma de las edades y puntuacion por ciudad y por Categoria
aggragated_data_multi = grouped_multi.agg(
    {
        'Edad': 'sum',
        'Puntuacion': 'mean'
    }
)
aggragated_data_multi
```

# Agregar datos a un Dataframe

```python
# Agregar una columna
# Crear un Dataframe de ejemplo

data = {'Nombre': ['Juan', 'Ana', 'Luis', 'Laura'],
        'Edad': [25,33,30,28]}

df = pd.DataFrame(data)
df
```

```python
# Agregar una columna
df['Ciudad'] = ['Acapulco','Tokyo', 'Nueva York', 'Barcelona']
df
```

```python
# Generar una nueva fila
new_row = pd.Series({'Nombre':'Pedro','Edad':45,'Ciudad':'Amsterdam'})

# Agregamos fila al Dataframe
df = pd.concat([df,new_row.to_frame().T], ignore_index=True)
df
```

### Combinando Dataframes

```python
# Crear un Dataframe de ejemplo

data1 = {'Nombre': ['Juan', 'Ana', 'Luis', 'Laura'],
        'Edad': [25,33,30,28]}

# Crear un segundo Dataframe de ejemplo
data2 = {'Nombre': ['Carla', 'Irene', 'Batman', 'Fulanito'],
        'Edad': [25,56,70,98]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df1['Ciudad'] = ['Acapulco','Tokyo', 'Nueva York', 'Barcelona']
df2['Ciudad'] = ['Acapulco','Merida', 'El Congo', 'Madrid']
```

```python
df_combine = pd.concat([df1,df2],ignore_index=True)
df_combine
```
