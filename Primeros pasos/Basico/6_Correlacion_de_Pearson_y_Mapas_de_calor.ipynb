# >>>> Correlacion <<<<

#### Magnitud y direccion de una relacion lineal entre dos variables

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Datos/personas.csv')
data
```

#### Relacion entre altura y peso

```python
plt.scatter(data['altura'],data['peso'])
plt.xlabel('Altura (cms)')
plt.ylabel('Peso (kgs)')
```

#### Relacion entre el Ingreso y las Horas trabajadas

```python
plt.scatter(data['ingreso'],data['horas_trabajadas'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Horas Trabajadas ()')
# La correlacion no nos dice nada con respecto a la causalidad
# Lo unico que nos indican es que tienen una relacion
```

#### Relacion entre Ingreso y Ausencia

```python
plt.scatter(data['ingreso'],data['ausencias'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Ausencias')
```

#### Relacion entre el Ingreso y el Peso de una persona

```python
plt.scatter(data['ingreso'],data['peso'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Peso (kgs)')
```

#### Matriz de correlacion

```python
data.corr() # -1 (existe una relacion fuerte con pendiente negativa) 0 1 (existe una relacion fuerte con pendiente positiva)
```

```python
matriz = data.corr()
plt.matshow(matriz, cmap='bwr',vmin=-1,vmax=1)
plt.xticks(range(5),data.columns, rotation= 45)
plt.yticks(range(5),data.columns)
for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i,j, round(matriz.iloc[i,j], 2), ha='center', va='center')



plt.colorbar()
```
